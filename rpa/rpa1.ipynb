```python
from openpyxl import load_workbook
from openpyxl.styles import Font, Border, PatternFill, Alignment
from openpyxl.styles.borders import Side

# 엑셀파일 열기
wb = load_workbook("./range.xlsx")
# 현재 활성화된 시트 가져오기
ws = wb.active
# A 컬럼 너비 조절
ws. column_dimensions["A"].width = 5
# 1 행 높이 조절
ws.row_dimensions[1].height = 50
# 셀 가져오기
a1 = ws["A1"]
b1 = ws["B1"]
c1 = ws["C1"]
# 폰트 속성 지정
a1.font = Font(color="ff0000", italic=True, bold=True)
b1.font = Font(color="cc33ff", name="arial", strike=True)
c1.font = Font(color="0000ff", size=20, underline="single")

# 테두리 생성
thin_border = Border(left=Side(style="thin"),right=Side(style="thin"),
                     top=Side(style="thin"),bottom=Side(style="thin"))
# 테두리 지정
a1.border = thin_border
b1.border = thin_border
c1.border = thin_border

for row in ws.rows:
    for cell in row:
        cell.alignment = Alignment(horizontal="center", vertical="center")
        if cell.column == 1:
            continue

        if isinstance(cell.value, int) and cell.value >80:
            cell.fill = PatternFill(fgColor="00ff00", fill_type="solid")
            cell.font = Font(color="ff0000")

wb.save("./range_style.xlsx")
```

```python
from openpyxl import Workbook
from openpyxl import load_workbook

wb = Workbook()
ws = wb.active

# 학번, 출석,퀴즈1,퀴즈2, 중간고사,기말고사,프로젝트
# 1,10,8,5,14,26,12
# 2,7,3,7,15,24,18
# 3,9,5,8,8,12,4
# 4,7,8,7,17,21,18
# 5,7,8,7,16,25,15
# 6,3,5,8,8,17,0
# 7,4,9,10,16,27,18
# 8,6,6,6,15,19,17
# 9,10,10,9,19,30,19
# 10,9,8,8,20,25,20
ws.append(["학번","출석","퀴즈1","퀴즈2", "중간고사","기말고사","프로젝트"])
scores =[
        [1,10,8,5,14,26,12,],
        [2,7,3,7,15,24,18,],
        [3,9,5,8,8,12,4],
        [4,7,8,7,17,21,18],
        [5,7,8,7,16,25,15],
        [6,3,5,8,8,17,0],
        [7,4,9,10,16,27,18],
        [8,6,6,6,15,19,17],
        [9,10,10,9,19,30,19],
        [10,9,8,8,20,25,20]
]

for score in scores:
    ws.append(score)


# scores.xlsx 생성
wb.save("./scores.xlsx")
```

```python
# scores.xlsx 읽은 후
wb = load_workbook("./scores.xlsx")
ws = wb.active
# 모든 학생들의 퀴즈2 점수 10 변경
# enumerate 인덱스 변경 가능
for idx, cell in enumerate(ws["D"]):
    # if cell.value == "퀴즈2":
    #     continue
    if idx == 0:
        continue
    cell.value = 10

# 새로운 칼럼추가
# h1 셀에 총점 I1 성적 추가한 후
ws["h1"].value = "총점"
ws["i1"].value = "성적"

# 총점 = 출석+퀴즈1+퀴즈2+중간고사+기말고사+프로젝스
for idx, score in enumerate(scores,start=2):
    ws.cell(row=idx,column=8).value = "=sum(b{0}:G{0})".format(idx)
    

    # 성적  
    # >=90 A,>=80 B,>=70 C, D, 단 출석점수가 5미만인 경우 F
    hap = sum(score[1:]) - score[3] + 10
    if hap >= 90:
        grade = "A"
    elif hap >= 80:
        grade = "B"
    elif hap >= 70:
        grade = "C"
    else:
        grade = "D"

    if score[1] < 5:
        grade = "F"

    ws.cell(row=idx, column=9).value = grade


wb.save("./scores.xlsx")
```
