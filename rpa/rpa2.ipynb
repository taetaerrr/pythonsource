```python
import pyautogui as p

size = p.size()
print(size)
print(size[0]) 
```

```python
# 마우스 조작

# 마우스 좌표 가져오기
position = p.position()
print(position)
print(position.x,position.y)
```

```python
# 마우스 이동 : moveTo(x,y,움직이는시간) - 절대좌표
p.moveTo(100,100,0.5)
p.moveTo(200,200,0.5)
p.moveTo(300,300,0.5)
```

```python
# 마우스 이동 : move(),moveREl() - 상대좌표
p.moveTo(100,100,0.5)
p.move(300,300,0.25)
print(p.position())
p.move(100,100,0.25)
p.moveRel
```
