#### 화면조작
- find_element(By.ID)
- find_element(By.CLASS_NAME)
- find_element(By.XPATH)
- find_element(By.CSS_SELECTOR)

- click() : 클릭
- send_keys() : 키 입력

```python
from chrome import set_chrome_driver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

driver = set_chrome_driver()

driver.get("https://www.naver.com")

# 검색창에 검색어 입력 후 엔터
# 검색 요소 찾기
element = driver.find_element(By.ID, "query")
# print(element)
element.send_keys("아이폰")
element.send_keys(Keys.ENTER)

# 검색 후 페이지 소스 가져오기
print(driver.page_source)
# 파싱 : selenium or BeautifulSoup
time.sleep(3)

driver.quit()
```

```python
# 스크립트 사용

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup


driver = webdriver.Chrome()
driver.maximize_window()

driver.get("https://www.naver.com/")
time.sleep(1)

# 두번째 탭 열기 : 스크립트 window.open() 사용
driver.execute_script("window.open('https://www.daum.net')")

# 브라우저 창에 열린 모든 창 정보 가져오기
tabs = driver.window_handles

# 다음으로 이동
driver.switch_to.window(tabs[1])
# 다음에서 검색어 입력 
element = driver.find_element(By.ID,"q")
element.send_keys("아이폰")
element.send_keys(Keys.ENTER)
time.sleep(1)

# 네이버에서 검색어 입력
driver.switch_to.window(tabs[0])
element = driver.find_element(By.ID,"query")
element.send_keys("안드로이드")
element.send_keys(Keys.ENTER)



time.sleep(2)

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup

driver = webdriver.Chrome()
driver.maximize_window()

driver.get("https://shopping.naver.com/ns/home")
time.sleep(1)

# 상품명 입력
element = driver.find_element(By.CLASS_NAME,"_searchInput_search_text_83jy9")
element.send_keys("마우스")
element.send_keys(Keys.ENTER)
time.sleep(2)

# 페이지 이동 후 스크롤 이동 : window.scrollTo(시작위치, 종료위치)
# driver.excute_script("window.scrollTo(0,2160)")

# 현재 문서 높이
prev_height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")

    time.sleep(2)

    # 스크롤이 진행된 후 문서 높이
    curr_height = driver.execute_script("return document.body.scrollHeight")

    if curr_height == prev_height:
        break

    prev_height = curr_height

# 맨 처음으로 가기
driver.execute_script("window.scrollTo(0,0)")

time.sleep(2)

driver.quit()
```
